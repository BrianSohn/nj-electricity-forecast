```python
import pandas as pd

# Path to your processed parquet
processed_path = "../data/processed/electricity_backfill_200101_202506.parquet"

# Load into a DataFrame
df = pd.read_parquet(processed_path)
```

```python
# Quick look at first and last rows
print(df.head())
print(df.tail())

# Check data types
print(df.dtypes)

# Check for missing values
print(df.isna().sum())

# Check overall shape
print(f"Total rows: {df.shape[0]}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(df['period'], df['sales'])
plt.title("NJ Residential Electricity Consumption (2001-2025)")
plt.xlabel("Month")
plt.ylabel("Sales (million kWh)")
plt.grid(True)
plt.show()
```

```python
# Check for duplicate months
duplicates = df['period'].duplicated().sum()
print(f"Duplicate periods: {duplicates}")

# Check for gaps in months
all_months = pd.date_range(df['period'].min(), df['period'].max(), freq='MS')
missing_months = all_months.difference(df['period'])
print(f"Missing months: {missing_months}")
```
